## Logical operators

Boolean operators can be combined using logical operators.

|Operator|Description|Example|
|--------|-----------|-------|
|and|Returns True if both statements are true|x < 5 and  x < 10|
|or|Returns True if one of the statements is true|x < 5 or x < 4|
|not|Reverse the result, returns False if the result is true|not(x < 5 and x < 10)|

```python
x = 4

print(x < 5 and  x < 4)
```

```python
print(x < 5 or x < 4)
```

```python
print(not(x < 5 and x < 4))
```

# Elif

The elif keyword mens 'else if' and is Python's way of saying "if the previous conditions were not true, then try this condition".

```python
a = 3
if a > 4:
  print('a is bigger than 4')
elif a > 2:
  print('a is bigger than 2')
```

This is different to using two if statetments, since using elif statemnts only run if the first if statement is False

```python
a = 5
if a > 4:
  print('a is bigger than 4')
if a > 2:
  print('a is bigger than 2')
```

# Else

The else keyword catches anything which isn't caught by the preceding conditions.

```python
a = 200
b = 33
if b > a:
  print("b is greater than a")
else:
  print("b is not greater than a")
```

```python
if b > a:
  print("b is greater than a")
elif a == b:
  print("a and b are equal")
else:
  print("a is greater than b")
```

## Nested if statements

You can have if statements inside if statements, this is called nested if statements.

```python
x = 41

if x > 10:
  print("Above ten,")
  if x > 20:
    print("and also above 20!")
  else:
    print("but not above 20.")
```

# Loops

In python, loops are used if you need to repeat the same piece of code several times. This repeated looping process is called 'iteration'.

A 'for' loop is used to iterate a fixed number of times over a sequence (such as a list or a string). They are defined using the keyword 'for' followed by a variable such as x, then the sequence to iterate over. In each iteration, the next item will be stored in the variable x until the end of the sequence is reached.

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
for x in "banana":
  print(x)
```

## Break

With the break statement we can stop the loop before it has looped through all the items.

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
```

## Range function

To loop through a set of code a specified number of times, we can use the <em>[range](https://www.w3schools.com/python/ref_func_range.asp)</em> function.

The <em>range</em> function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.

```python
for x in range(6):
  print(x, 'Hello :)')
```

The <em>range</em> function defaults to 0 as a starting value, however it is possible to specify the starting value by adding a parameter.

```python
for x in range(2, 6):
  print(x)
```

The <em>range</em> function defaults to increment the sequence by 1, however it is possible to specify the increment value by adding a third parameter.

```python
for x in range(2, 30, 3):
  print(x)
```

## Nested loops
A nested loop is a loop inside a loop.

The "inner loop" will be executed one time for each iteration of the "outer loop".

```python
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for x in adj:
  for y in fruits:
    print(x, y)
```
