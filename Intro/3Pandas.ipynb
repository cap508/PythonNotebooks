# Series

A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type. The code below turns a list into a series.

```python
a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

We can find specific values in a series by indexing just like with lists

```python
print(myvar[0])
```

# Dataframes

Data sets in Pandas are usually multi-dimensional tables, called DataFrames.

Series is like a column, a DataFrame is the whole table.

```python
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data)

print(df)
```

As you can see from the result above, the DataFrame is like a table with rows and columns.

Pandas use the loc attribute to return one or more specified row(s)

```python
print(df.loc[0])
```

```python
print(df.loc[[0, 1]])
```

When returning just one row we get a Series, when returning multiple rows we get a DataFrame

# CSV Files

A simple way to store big data sets is to use CSV files (comma separated files).

CSV files contains plain text and is a well know format that can be read by everyone including Pandas.

In our examples we will be using a CSV file with data used for lung cancer identification.

```python
df = pd.read_csv('./Data/survey lung cancer.csv')
```

We can look at the first 5 rows of this file by using the function [head](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html).

```python
df.head()
```

The [shape](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shape.html) function can be used to see how many rows and columns a DataFrame has

```python
print(df.shape)
```

This means that the DataFrame has 309 rows and 16 columns.


We can use the [info](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.info.html) function to get more information about the dataset.

```python
print(df.info())
```

This tells us the number of rows and columns, the name of each column, and the data type of each colummn.

All the columns except for GENDER, and LUNG_CANCER are integers.

Lets look more specifically at some of the columns.

The function [value_counts](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html) will tell you how many of each value are in each column. To select a specific column we put the column name in speech marks within square brackets.


```python
print(df['GENDER'].value_counts())
```

```python
print(df['SMOKING'].value_counts())
```

```python
print(df['AGE'].value_counts())
```

For most of the columns using value_counts is a good way of looking at the information because there are only two values in that columns. For age value counts doesn't tell us much because there are lots of different values.

Instead knowing the maximum, minimum, and mean age may be more helpful.

Getting the minimum and maximum of a column is easy

```python
print(min(df['AGE']))
print(max(df['AGE']))
```

To get the mean we must do some more work. The mean of a set of numbers is the sum of the numbers divided by the length of the set of numbers.

We could get the mean in a number of ways. One way would be to use the sum and len functions. The sum function adds all the values in a series, and the len function finds the length of a series.

```python
print(sum(df['AGE'])/len(df['AGE']))
```
