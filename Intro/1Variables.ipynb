# Variables

Variables contain data. To create a variable simply choose a variable name and assign it a value. There are a few rules for variable names:
- A variable name must start with a letter or the underscore character
- A variable name cannot start with a number
- A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
- Variable names are case-sensitive (age, Age and AGE are three different variables)
- A variable name cannot be any of the Python [keywords](https://www.w3schools.com/python/python_ref_keywords.asp).

When you run this cell you won't see anything. This is because the code is simply assigning variables a value.

```python
y = 1
x = 'word'
```

In the above cell we create two variables called y and x.

y is a variable with the value 1

x is a variable with the value 'word'

There are many different types of variables:
- string - usually words are created by putting ' ' or " " around a value
- integer - a whole number created by simply typing a number
- float - a decimal number created by simply typing a number
- boolean - True or False created by typing True or False (note the capital letter at the start)

```python
string_variable = 'hello world!'
integer_variable = 3
float_variable = 1.5
boolean_variable = True
```

## Changing variable types

In python changing variable types is easy (mostly!)

|Variable type to change to|Function|
|--------------------------|--------|
|String                    |str()   |
|Integer                   |int()   |
|Float                     |float() |

```python
new_string_variable = '5'
change_to_integer = int(new_string_variable)
change_to_float = float(new_string_variable)
```

## Displaying variables

To see what a variable looks like we can use the <em>print</em> function

```python
print(new_string_variable)
print(change_to_integer)
print(change_to_float)
```

## Checking a variable type

The string and integer variable look the same when we print them so how do we know what type of variable they are? We use the <em>type</em> function.

```python
print(type(new_string_variable))
print(type(change_to_integer))
print(type(change_to_float))
```

# Lists

Lists are used to store multiple items in a single variable. Lists are created using square brackets:

```python
first_list = ["apple", "banana", "cherry"]
```

Above is a list of strings. You can make lists of any data type (and even mixed data types). We can easily find out how many things are in a list using the <em>len</em> function.

```python
length_of_first_list = len(first_list)
print(length_of_first_list)
```

You can find specific items in lists using indexing. This is done by writing the list you want to get and item from and then the number of the item you want in square brackets. In python you start counting at 0.

```python
first_item_in_list = first_list[0]
print(first_item_in_list)

second_item_in_list = first_list[1]
print(second_item_in_list)

last_item_in_list = first_list[-1]
print(last_item_in_list)

first_and_second_item = first_list[:2]
print(first_and_second_item)
```

# Dictionaries

Dictionaries are used to store data values in key:value pairs where the key is used to look up the corresponding value.

Dictionarys are defined using curly brackets, with key:value pairs separeted by commas:

```python
costs = {'apple':0.82,
         'banana':0.95,
         'orange':1.51,
         'kiwi':1.25,
         }
```

To access the corresponding value associated with each key, we use square brackets, just like in lists. However, in Python dictionarys have no order, so we cant look up items using indexing. e.g. to find the cost of a banana, we could use:

```python
print(costs['banana'])
```

```python
cost_apple = costs['apple']
cost_5_kiwis = costs['kiwi']*5
cost_2_oranges_and_2_bananas = costs['orange']*2 + costs['banana']*2
```

```python
#@title
if cost_apple == costs['apple']:
  print("That's correct, the cost of an apple is Â£{}!".format(cost_apple))
else:
  print("Sorry thats not quite right for the cost of an apple! Have another go")

if cost_5_kiwis == 5*costs['kiwi']:
  print("That's correct, the cost of 5x kiwis is Â£{}!".format(cost_5_kiwis))
else:
  print("Sorry thats not quite right for the cost of 5x kiwis! Have another go")

if cost_2_oranges_and_2_bananas == 2*costs['orange'] + 2*costs['banana']:
  print("That's correct, the cost of 2x orange and 2x banana is Â£{}!".format(cost_2_oranges_and_2_bananas))
else:
  print("Sorry thats not quite right for the cost of 2x orange and 2x banana! Have another go")
```
