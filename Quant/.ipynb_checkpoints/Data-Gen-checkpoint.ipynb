```python
def withinStd(data, mul):
    limitU = data.mean() + mul*data.std()
    limitL = data.mean() - mul*data.std()
    countU = sum(map(lambda x:x>limitU, data))
    countL = sum(map(lambda x:x<limitL, data))

    return(1 - (countU+countL)/dataSize)

print(withinStd(dataNormal, 1))
print(withinStd(dataNormal, 2))
```

```python
# Calculate the confidence interval for a mean from a sample


import numpy as np
import scipy.stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h


dataSize = 30
data = np.random.normal(loc=0.0, scale = 2.0, size = dataSize)

ret = mean_confidence_interval(data)
print(ret)


# Check this?


count = 0
for i in range(100):
    data = np.random.normal(loc=0.0, scale = 2.0, size = dataSize)
    m = np.mean(data)
    if m < ret[1]:
        count = count +1
    if m > ret[2]:
        count = count+1

print (f"Count {count}")



```

```python
dataSize = 10000
treat1 = np.random.normal(loc=1.5, scale = 1.0, size = dataSize)
treat2 = np.random.normal(loc=2.5, scale = 1.0, size = dataSize)

np.savetxt("treat1.csv", treat1, delimiter=',')
np.savetxt("treat2.csv", treat2, delimiter=',')


exp1 = random.choice(treat1, 30, replace = False)
exp2 = random.choice(treat2, 30, replace = False)
```

# Two sample T-Test

```python
import pandas as pd
import scipy.stats as stats
```

```python
# first test if the two groups have the same variance
# https://www.geeksforgeeks.org/how-to-conduct-a-two-sample-t-test-in-python/
# https://www.jmp.com/en_gb/statistics-knowledge-portal/t-test/two-sample-t-test.html

print(np.var(exp1), np.var(exp2))
```

```python
stats.ttest_ind(a=exp1, b=exp2, equal_var=True)
```

```python
#Visualize

plt.hist(exp1, alpha=0.5, color='red')
plt.hist(exp2, alpha=0.5, color='blue')
plt.show()
```

# One Sample T-Test

```python
# https://www.jmp.com/en_gb/statistics-knowledge-portal/t-test/one-sample-t-test.html
```

```python
# is the average reduction EQUAL to 1.5
exp1 = random.choice(treat1, 30, replace = False)
print(np.mean(exp1))
stats.ttest_1samp(exp1, popmean=1.5)
```

```python
# is the average reduction Greater than 1
exp1 = random.choice(treat1, 30, replace = False)
print(np.mean(exp1))
stats.ttest_1samp(exp1, popmean=1.0, alternative='greater' )
```

```python
# https://www.reneshbedre.com/blog/mann-whitney-u-test.html


import pandas as pd
df = pd.read_csv("https://reneshbedre.github.io/assets/posts/mann_whitney/genotype.csv")
df.head(2)

# get summary statistics
df.agg(["count", "min", "max", "median", "mean", "skew"])

# generate boxplot to check data spread
import matplotlib.pyplot as plt
df.boxplot(column=['A', 'B'], grid=False)
plt.show()
```

```python
import scipy.stats as stats
w, pvalue = stats.shapiro(df['A'])
print(w, pvalue) # p-value < 0.05 == not normal

w, pvalue = stats.shapiro(df['B'])
print(w, pvalue)

# plot histogram
import matplotlib.pyplot as plt
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Frequency histogram of genotypes yield')
ax1.hist(df['A'], bins=10, histtype='bar', ec='k')
ax2.hist(df['B'], bins=10, histtype='bar', ec='k')
ax1.set_xlabel("Yield")
ax2.set_xlabel("Yield")
plt.show()
```

```python
import scipy.stats as stats
# perform two-sided test. You can use 'greater' or 'less' for one-sided test
stats.mannwhitneyu(x=df['A'], y=df['B'], alternative = 'two-sided')
```

```python
import scipy.stats as stats
stats.mannwhitneyu(x=df['A'], y=df['B'], alternative = 'greater')
```
